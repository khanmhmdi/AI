```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

torch.random.manual_seed(0)
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)

tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")

messages = [
    {"role": "system", "content": "You are a helpful AI assistant."},
    {"role": "user", "content": "Can you provide ways to eat combinations of bananas and dragonfruits?"},
    {"role": "assistant", "content": "Sure! Here are some ways to eat bananas and dragonfruits together: 1. Banana and dragonfruit smoothie: Blend bananas and dragonfruits together with some milk and honey. 2. Banana and dragonfruit salad: Mix sliced bananas and dragonfruits together with some lemon juice and honey."},
    {"role": "user", "content": "What about solving an 2x + 3 = 7 equation?"},
]

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
)

generation_args = {
    "max_new_tokens": 500,
    "return_full_text": False,
    "temperature": 0.0,
    "do_sample": False,
}

output = pipe(messages, **generation_args)
print(output[0]['generated_text'])
```

```python
class PlacementModel(torch.nn.Module):
    def __init__(self, num_experts=10, num_devices=5, embedding_dim=32):
        super().__init__()
        self.expert_embed = torch.nn.Embedding(num_experts, embedding_dim)
        self.device_embed = torch.nn.Embedding(num_devices, embedding_dim)

        self.fc = torch.nn.Sequential(
            torch.nn.Linear(embedding_dim*2 + 2, 128),  # 2*emb + capacity + speed
            torch.nn.ReLU(),
            torch.nn.Linear(128, num_devices)
        )

    def forward(self, expert_ids, device_capacities, speed_matrix):
        # Expert embeddings
        expert_emb = self.expert_embed(expert_ids)

        # Device context embedding
        device_ids = torch.arange(len(device_capacities))
        device_emb = self.device_embed(device_ids)

        # Combine features
        batch_size = expert_emb.size(0)
        device_features = torch.cat([
            device_emb,
            device_capacities.unsqueeze(-1),
            speed_matrix.mean(dim=1).unsqueeze(-1)  # Average connection speed
        ], dim=1)

        # Expand device features for each expert
        combined = torch.cat([
            expert_emb,
            device_features.repeat(expert_emb.size(0), 1)
        ], dim=1)

        return torch.softmax(self.fc(combined), dim=-1)
```
