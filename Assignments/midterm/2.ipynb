```python
m = int(input()) #number of regions
b = int(input()) #number of bids
c = int(input()) #number of operators
```

```python
offers = list()
local_maximum = 0
result = None

for _ in range(b):
    offer_details = list(map(float, input()[: -1].split()))
    protocol_name = int(offer_details[0])
    price = offer_details[1]
    regions = list(map(int, offer_details[2:]))
    offers.append([protocol_name, price, regions])
```

```python
def get_state_value(state):
    summation = 0
    for i in range(b):
        if state[i] == 1:
            income = offers[i][1]
            regions = offers[i][2]
            summation += income
    return summation

def get_random_neighbor(state):
    global local_maximum
    random_no = random.randint(0, b)
    if state[random_no] == 1:
        state[random_no] = 0
    else:
        state[random_no] = 1
    return state

def local_search(state):
    random_state = get_random_neighbor(state.copy())
    random_result = get_state_value(random_state.copy())
#     condition = random_result > local_maximum
    if random_result > local_maximum:
        result = random_state.copy()
        local_maximum = random_result
        local_search(result)
        

initial_state = []
for i in range(b):
    initial_state.append(0)

local_search(initial_state.copy())

for i in range(5000):
    initial_state = result[:]
    local_search(initial_state[:])

print(result)
```
